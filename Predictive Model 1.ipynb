```python
import pandas as pd
```

```python
df=pd.read_csv('/content/bike_sales.txt')
```

```python
df['Profit'] = df['Profit'].str.replace('$', '', regex=False)
df['Profit'] = df['Profit'].str.replace(' ', '', regex=False)
df['Profit'] = df['Profit'].str.replace(',', '', regex=False)
```

```python
df['Cost'] = df['Cost'].str.replace('$', '', regex=False)
df['Cost'] = df['Cost'].str.replace(' ', '', regex=False)
df['Cost'] = df['Cost'].str.replace(',', '', regex=False)
```

```python
df['Profit'] = pd.to_numeric(df['Profit'])
df['Cost'] = pd.to_numeric(df['Cost'])
```
